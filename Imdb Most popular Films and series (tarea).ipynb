```python
# CONDE AJURURO CRISTHIAN CI:9895070
# Importando las librerias en Python
import numpy as np
import pandas as pd     
```

```python
# Importación de datos de la base de datos:
dataset = pd.read_csv('/content/drive/MyDrive/IMDb.csv')
dataset
```

```python
dataset.info()
#muchas de las columnas son de tipo object, estas hay q convertirlas a tipo numerico
```

```python
#cambiamos la columna 'Rate' a tipo numerico
dataset['Rate'] = pd.to_numeric(dataset['Rate'], errors = 'coerce') 
dataset[dataset['Rate'].isnull()]
```

```python
#Remplazamos los valores nulos por 0
dataset.fillna(0, inplace = True) 
dataset
```

```python
#Hacemos lo mismo para las colunmas Votes, Duration, and Episodes
dataset['Votes'] = dataset['Votes'].replace(",", "", regex=True)
dataset['Votes'] = pd.to_numeric(dataset['Votes'], errors = 'coerce')
dataset['Duration'] = pd.to_numeric(dataset['Duration'], errors = 'coerce')
dataset['Episodes'] = pd.to_numeric(dataset['Episodes'], errors = 'coerce')
dataset.fillna(0, inplace = True)
dataset
```

```python
#cambiamos la columna  a int
dataset['Votes'] = dataset['Votes'].astype('int')
dataset['Duration'] = dataset['Duration'].astype('int')
dataset['Episodes'] = dataset['Episodes'].astype('int')
dataset
```

```python
#encontramos los valores unicos del dataset
dataset['Frightening'].unique()
```

```python
#Categorizamos por nivel los valores siendo 0 'Sin' o 'NO calificado', Mild 1, moderado 2 y severo a 3 
dataset = dataset.replace({'None':0,'Mild':1,'Moderate':2,'Severe':3,'No Rate':0})
dataset
```

```python
#eliminamos los datos duplicados
dataset = dataset.drop_duplicates(keep = 'first')
dataset
```

```python
print(dataset.describe())
print(dataset.info())
```

3.1. Calificación média (Rate).

```python
C_media=np.median(dataset['Rate'])
print("la calificacion media de las peliculas y series en IMDB es ")
C_media
```

3.2. Total de Votos por género.


```python
Total_v = dataset.groupby(['Genre'])['Votes'].sum()
Total_v
```

3.3. Duración media de películas y series, e.g. Sumar la columna de Duration de películas y dividirlo por el total de películas.

```python
D_media=np.sum(dataset['Duration'])/5028
print("la duracion media de las peliculas y series en IMDB es ")
D_media
```

3.4. Porcentaje de películas clasificadas por Violencia o Violence (5 categorias).


```python
p_violentas=(np.count_nonzero(dataset['Violence'])/5028) *100
print("el porcentaje peliculas y series con calificiacion violenta en IMDB es ")
p_violentas
```

3.5. Porcentaje de Series clasificadas por ser Atemorizante o Frightening (5 categorias).




```python
p_terror=(np.count_nonzero(dataset['Frightening'])/5028) *100
print("el porcentaje peliculas y series con calificiacion Atemorizante en IMDB es ")
p_terror
```

3.6. Año en que hubo más series violentas calificadas como severas (severe)





```python
año_violento  = dataset[(dataset['Violence'] == 3) & (dataset['Type'] == 'Series')]
print("Se muestra el año seguida de la cantidad de series mas violentas")
año_violento.groupby(['Date'])['Violence'].value_counts().idxmax()
```

3.7. Año en que hubo más películas con escenas de alcohol (Alcohol).



```python
año_alcohol  = dataset[(dataset['Alcohol'])& (dataset['Type'] == 'Film')]
print("Se muestra el año seguida de la cantidad de peliculas con escenas de alcohol")
año_alcohol.groupby(['Date'])['Violence'].value_counts().idxmax()
```

3.8. Lista de películas clasificadas como Severe en las 5 categorias (Nudity, Violence, Profanity, Alcohol, Frightening)

```python
lista_p = dataset[(dataset['Nudity'] == 3) & (dataset['Violence'] ==3) & (dataset['Profanity'] == 3) & (dataset['Alcohol'] == 3) & (dataset['Frightening'] == 3) ]
lista_p = lista_p[['Name', 'Date', 'Rate','Genre', 'Type']]
print("Lista de peliculas y series con calificiacion 'Severe' ")
lista_p
```

3.9. Obtener la lista de las mejores 10 películas según IMDb.


```python
lista_p1 = dataset[(dataset['Rate'] > 8.7) & (dataset['Type'] =='Film') ]
print("Lista de las mejores 10 peliculas segun IMDB ")
lista_p1 = lista_p1[['Name', 'Date', 'Rate','Genre', 'Type']]
lista_p1.head(10)
```

3.10. Películas con más de 1 millón de votos ordenados por su ranking IMDB.

CONDE AJURURO CRISTHIAN CI:9895070



```python
lista_p2 = dataset[(dataset['Votes'] >1000000) ]
print("Lista de peliculas y series con más de 1 millón de votos ordenados por su ranking  ")
lista_p2 = lista_p2[['Name', 'Date', 'Rate','Genre', 'Type','Votes']]
lista_p2=lista_p2.sort_values(by = 'Rate', ascending = False)
lista_p2.head(10)
```
