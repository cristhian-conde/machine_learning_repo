```python
#Importando librerias
from sklearn import datasets,linear_model
```

```python
#Importando los datos
boston=datasets.load_boston()
print(boston)
```

```python
#Verificar la información del dataset
print(boston.keys())
```

```python
#Caracteristicas del dataset
print(boston.DESCR)
```

```python
#Ver la cantidd de datos
print(boston.data.shape)
```

```python
#nombre de las columnas
print(boston.feature_names)
```

```python
# Regresión lineal múltiple
#Seleccionar las columnsa 5,6,7 del dataset
x_multiple=boston.data[:,5:8]
x_multiple
```

```python
# Definimos los datos correspondientes a y
y_multiple=boston.target
```

```python
#implementando la regresión
from sklearn.model_selection import train_test_split

#Separando los datos de train en entrenamiento y prueba
x_train,x_test,y_train,y_test=train_test_split(x_multiple,y_multiple,test_size=0.3)
```

```python
#definimos el algoritmo a utilizar
lr_multiple=linear_model.LinearRegression()

#entrenando al modelo
lr_multiple.fit(x_train,y_train)
```

```python
#Realizola predicción
y_pred_multiple=lr_multiple.predict(x_test)
```

```python
#Datos del modelo
print('valor de la pendiente o coeficiente')
print(lr_multiple.coef_)
```

```python
#Valor de la intersección
print('Valor de la intersección')
print(lr_multiple.intercept_)
```

```python
print('Precisión del modelo')
print(lr_multiple.score(x_train,y_train))
```

```python
#Calculando RMSE
import math
from sklearn import  metrics
MSE=metrics.mean_squared_error(y_test,y_pred_multiple)
print('RMSE=%.02f'%math.sqrt(MSE))
```
