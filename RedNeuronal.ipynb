```python
import numpy as np
import pandas as pd
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
```

```python
#Cargando el archivo
data=pd.read_csv('/content/drive/MyDrive/Copia de default.csv')
data
```

```python
# Reconfigurando la estructura convirtiendo texto en  número
data=data.replace(to_replace='Yes', value=1)
data=data.replace(to_replace='No',value=0)
data
```

```python
# Reconfigurando la estructura del data
y=data['default'].values.reshape(-1,1)
x=data[['balance','income','student']].values.reshape(-1,3)

print('x=',x.shape)
print('y=',y.shape)
```

```python
#realizar el muestreo : train=70%  test=30%
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.3,shuffle=True,random_state=10)

print('x_train=', train_x.shape)
print('y_train=', train_y.shape)
print('x_test=', test_x.shape)
print('y_test=', test_y.shape)
train_x
```

```python
#Creamos nuestra red neuronal
ann = MLPClassifier(activation="logistic", solver="sgd", hidden_layer_sizes=(10, 200, 100, 100, 40), random_state=1)
ann.fit(train_x,train_y)
predicciones=ann.predict(test_x)
predicciones
```

```python
#Calculamos la matriz de confusión
from sklearn.metrics import confusion_matrix
tp,fp,fn,tn=confusion_matrix(test_y,predicciones,labels=[1,0]).ravel()

print('TN=',tn)
print('FP=',fp)
print('FN=',fn)
print('TP',tp)
print()
confusion_matrix(test_y,predicciones,labels=[1,0])
```

```python
#Calculamos el Accurracy
acc=(tn+tp)/(tn+fp+fn+tp)
print('Accurracy:'+str(acc))
```

```python
#Calculanco la sensitividad del modelo
sencitivity=tp/(tp+fn)
print('Sencitivity:'+str(sencitivity))
```

```python
#Calculanco la Especificidad del modelo
specificity=tn/(tn+fp)
print('Specificity:'+str(specificity))
```
