```python
# Importando las librerias en Python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
import matplotlib.pyplot as plt
import numpy as np
```

```python
x1 = np.array([3,1,1,2,1,6,6,6,5,6,7,8,9,8,9,9,8])
x2 = np.array([5,4,5,6,5,8,6,7,6,7,1,2,1,2,3,2,3])
plt.plot()
plt.xlim([0,10])
plt.ylim([0,10])
plt.title("conjunto de datos")
plt.scatter(x1,x2)
plt.show()
```

```python
X=np.array(list(zip(x1,x2))).reshape(len(x1),2)
```

```python
grupos =[]
K =range(1,10)
for k in K:
  Modelokmean = KMeans(n_clusters=k).fit(X)
  Modelokmean.fit(X)
  grupos.append(sum(np.min(cdist(X, Modelokmean.cluster_centers_,'euclidean'),
                    axis=1))/X.shape[0])

plt.plot(K,grupos,'bx-')
plt.xlabel('Clusteres')
plt.ylabel('Suma total de errores cuadraticos')
plt.title('metodo codo')
plt.show()
```
