```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('/content/drive/MyDrive/Precio_viviendas_Australia.csv')
data
```

```python
data.info()
```

```python
data.describe()
```

```python
#cambiamos la columna 'habitaciones' a tipo numerico
data['Habitaciones'] = pd.to_numeric(data['Habitaciones'], errors = 'coerce') 
data
```

```python
#hacemos lo mismo con las siguientes columnas
data['Precio'] = pd.to_numeric(data['Precio'], errors = 'coerce') 
data['Distancia'] = pd.to_numeric(data['Distancia'], errors = 'coerce') 
data['Semana'] = pd.to_numeric(data['Semana'], errors = 'coerce') 
data['CPI'] = pd.to_numeric(data['CPI'], errors = 'coerce')
data['Ultimo_CPI'] = pd.to_numeric(data['Ultimo_CPI'], errors = 'coerce') 
data['Distancia2'] = pd.to_numeric(data['Distancia2'], errors = 'coerce') 
```

```python
data.info()
data.isnull().sum()
```

```python
#reemplazamos los valores nulos con 0 y convertimos a valor entero las siguientes colunmas
data.fillna(0, inplace = True)
data['Semana'] = data['Semana'].astype('int')
data['Precio'] = data['Precio'].astype('int')
data['Habitaciones'] = data['Habitaciones'].astype('int')
data
```

```python
data.info()
```

```python
#Extraemos del dataset los datos que nos serviran para ingresarlos en el metodo de regresion lineal
X = data.drop(['Suburbio', 'Direccion','Tipo','Metodo', 'VendedorG','Fecha','Precio'], axis=1)
X
```

```python
#El get_dummiesmétodo de la biblioteca de pandas convierte columnas categóricas
#en columnas numéricas. Luego, .iloc[:,1:]ignora la primera columna y devuelve 
#el resto de las columnas,esto se debe a que 
#siempre podemos representar "n" categorías con "n - 1" columnas.
#en este caso convertiremos las columnas "metodo" y "tipo"
metodo = pd.get_dummies(data.Metodo).iloc[:,1:]
metodo
```

```python
tipo = pd.get_dummies(data.Tipo).iloc[:,1:]
tipo
```

```python
#concatenamos las nuevas columnas a nuestro X 
X = pd.concat([tipo,X,metodo], axis=1)
X
```

```python
#Extramos del dataset la columna objetivo, esta sera nuestra y 
y= data.get(['Precio'])
y
```

```python
#implementando la regresión
from sklearn.model_selection import train_test_split
from sklearn import linear_model
#Separando los datos de train en entrenamiento y prueba
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
x_train
```

```python
#definimos el algoritmo a utilizar
lr_multiple=linear_model.LinearRegression()

#entrenando al modelo
lr_multiple.fit(x_train,y_train)
```

```python
#Realizola predicción
y_pred_multiple=lr_multiple.predict(x_test)
```

```python
#Datos del modelo
print('valor de la pendiente o coeficiente')
print(lr_multiple.coef_)
```

```python
#Valor de la intersección
print('Valor de la intersección')
print(lr_multiple.intercept_)
```

```python
print('Precisión del modelo')
print(lr_multiple.score(x_train,y_train))
```

```python
#Calculando RMSE
import math
from sklearn import  metrics
MSE=metrics.mean_squared_error(y_test,y_pred_multiple)
print('RMSE=%.02f'%math.sqrt(MSE))
```
